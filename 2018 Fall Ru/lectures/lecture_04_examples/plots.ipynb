# Plots

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

# InteractiveShell.ast_node_interactivity = 'none'
# %config InlineBackend.figure_format = 'svg'

%matplotlib inline
# %pylab inline
plt.rcParams['figure.figsize'] = (10.0, 4)

import numpy as np
import scipy as sp
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x))

plt.plot(x, np.cos(x))

plt.legend(['sin', 'cos'], loc='upper right')
```

```python
x = np.linspace(0, 5, 20)

plt.subplot(1,2,1)
plt.title('sin')
plt.plot(x, np.sin(x))

plt.subplot(1,2,2)
plt.title('cos')
plt.plot(x, np.cos(x), 'b-', x, np.cos(x), 'ro')
```

```python
t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('x = 2', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='red', shrink=0.1),
            )

plt.ylim(-2,2)
plt.show()
```

```python
x = np.arange(10) + 1
y = x

plt.subplot(121)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')

plt.subplot(122)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
```

```python
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.001)
plt.title('symlog')
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 

plt.scatter(x, y, s=area, c=colors, cmap='inferno', alpha=0.5,)

plt.title('Scatter', fontweight='bold', fontsize='x-large')
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

plt.hist(x, 50, normed=1, facecolor='b')

plt.xlabel('random variable value')
plt.ylabel('probability')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])

plt.title('Hist', fontweight='bold', fontsize='x-large')
```

```python
N = 5
red = (20, 35, 30, 35, 27)
blue = (25, 32, 34, 20, 25)
red_sdt = (2, 3, 4, 1, 2)
blue_std = (3, 5, 2, 3, 3)
ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind, red, width, color='r', yerr=red_sdt)
p2 = plt.bar(ind, blue, width, color='b', bottom=red, yerr=blue_std)

plt.ylabel('Bar')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Red', 'Blue'))
```

```python
days = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]
plt.xlabel('Day')
plt.ylabel('Degrees Celsius')
plt.plot(days, celsius_min,
         days, celsius_min, "oy",
         days, celsius_max, 
         days, celsius_max, "or")
print("The current limits for the axes are:")        
print(plt.axis())
print("We set the axes to the following values:")
xmin, xmax, ymin, ymax = 0, 10, 14, 45
print(xmin, xmax, ymin, ymax)
plt.axis([xmin, xmax, ymin, ymax])
plt.show()
```

```python
X = np.linspace(-2 * np.pi, 2 * np.pi, 50, endpoint=True)
F1 = 3 * np.sin(X)
F2 = np.sin(2*X)
F3 = 0.3 * np.sin(X)
startx, endx = -2 * np.pi - 0.1, 2*np.pi + 0.1
starty, endy = -3.1, 3.1
plt.axis([startx, endx, starty, endy])
plt.plot(X,F1)
plt.plot(X,F2)
plt.plot(X,F3)
plt.plot(X, F1, 'ro')
plt.plot(X, F2, 'bx')
plt.show()
```
